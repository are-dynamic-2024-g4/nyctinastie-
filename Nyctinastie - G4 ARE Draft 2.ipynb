# Nyctinastie des plantes

Bienvenue dans ce Jupyter Notebook consacré à la modélisation de la nyctinastie des plantes. La nyctinastie est un phénomène biologique fascinant observé chez de nombreuses espèces végétales, où les feuilles se replient ou se ferment en réponse à des stimuli externes tels que les changements de lumière ou la variation de température, généralement observée la nuit. Ce mouvement est souvent associé à des avantages adaptatifs, tels que la protection contre les prédateurs, la réduction de la perte d'eau ou encore l'optimisation de l'absorption de la lumière.

Dans ce notebook, nous allons explorer différents aspects de la nyctinastie des plantes, y compris ses mécanismes sous-jacents, les facteurs environnementaux qui influent sur ce phénomène, ainsi que la manière dont il peut être modélisé mathématiquement. Nous aborderons également des exemples de modèles existants et nous allons implémenter notre propre modèle de nyctinastie à l'aide de Python et des bibliothèques scientifiques telles que Turtle, Matplotlib et NumPy.

## 1. Lotus sacré (Nelumbo nucifera)

La nyctinastie chez Nelumbo nucifera, plus communément appelé lotus sacré, repose sur le mécanisme de thermorégularisation. Ce processus permet à la plante de réguler son mouvement nyctinastique en réponse aux variations de température dans son environnement. Dans cette section, nous aborderons la modélisation de ce phénomène en nous basant sur la corrélation des données climatiques d'une région donnée.

### Mécanismes sous-jacents de la nyctinastie chez Nelumbo nucifera
Nelumbo nucifera utilise la thermorégularisation pour ajuster son mouvement nyctinastique en fonction des changements de température. Ce mécanisme lui permet de maintenir un équilibre thermique optimal, crucial pour sa survie.

### Période de la thermorégularisation

Cycle végétal du lotus sacré :

- **Phase de développement :** Cette phase comprend les étapes juvénile, végétative et reproductive de la plante. Pendant cette période, la plante se développe et accumule les ressources nécessaires à la floraison ultérieure.
- **Période de floraison :** C'est la période où la plante produit des fleurs.


Processus de floraison :

- **Induction florale :** Cette phase commence au printemps (mars - avril), où les températures de l'eau commencent à augmenter après l'hiver, incitant la plante à démarrer le processus de floraison.
- **Évocation et initiation florale :** Cela se produit au début de l'été (mai - juin), lorsque les conditions environnementales sont favorables à l'initiation des bourgeons floraux.
- **Développement floral :** Cette phase s'étend de la mi-été (juin) jusqu'à la fin de l'été (septembre). C'est pendant cette période que la nyctinastie, le mouvement de fermeture des fleurs la nuit, peut être modélisée. Ce phénomène peut être observé comme une réponse protectrice de la plante aux conditions nocturnes.

Températures :
- **Idéal :** 20°C à 35°C
- **Tolérable:** 0°C à 20°C & 35°C à 38°C
- **Inadéquat:** -10°C à 0°C & +38°C

```python
# Imputation des températures manquantes par complétion de série
# Interpolation linéaire des taux d'ouvertures en fonction de températures

# Fonction pour trouver la plage de température dans les données fournies
def trouver_plage_temperature(data):
    min_temp = min(data, key=lambda x: x[0])[0]  # Trouve la température minimale
    max_temp = max(data, key=lambda x: x[0])[0]  # Trouve la température maximale
    return list(range(min_temp, max_temp + 1))  # Retourne une liste de toutes les températures possibles dans les données

# Fonction pour créer un dictionnaire de température à ouverture à partir des données
def creer_dictionnaire_temp_to_opening(data):
    temp_to_opening = {temp: [] for temp, _ in data}  # Initialise un dictionnaire vide avec toutes les températures possibles
    for temp, opening in data:
        temp_to_opening[temp].append(opening)  # Ajoute les valeurs d'ouverture associées à chaque température dans le dictionnaire
    return temp_to_opening

# Fonction pour remplir les températures manquantes en utilisant une interpolation linéaire
def remplir_temp_manquantes(plage_temp, temp_a_ouverture, min_temp, max_temp):
    nouvelles_donnees = []
    for temp in plage_temp:
        if temp in temp_a_ouverture and temp_a_ouverture[temp]:
            moyenne_ouverture = sum(temp_a_ouverture[temp]) / len(temp_a_ouverture[temp])  # Calcul de la moyenne des ouvertures disponibles pour cette température
            nouvelles_donnees.append([temp, moyenne_ouverture])  # Ajoute la température et la moyenne d'ouverture correspondante aux nouvelles données
        else:
            temp_inferieure = next((t for t in range(temp, min_temp - 1, -1) if t in temp_a_ouverture and temp_a_ouverture[t]), None)  # Trouve la température la plus proche inférieure avec des valeurs d'ouverture disponibles
            temp_superieure = next((t for t in range(temp, max_temp + 1) if t in temp_a_ouverture and temp_a_ouverture[t]), None)  # Trouve la température la plus proche supérieure avec des valeurs d'ouverture disponibles
            if temp_inferieure and temp_superieure:
                ouverture_inferieure = temp_a_ouverture[temp_inferieure][-1]  # Dernière valeur d'ouverture à la température inférieure
                ouverture_superieure = temp_a_ouverture[temp_superieure][0]  # Première valeur d'ouverture à la température supérieure
                poids_inferieur = temp - temp_inferieure  # Poids relatif de la température inférieure
                poids_superieur = temp_superieure - temp  # Poids relatif de la température supérieure
                poids_total = poids_inferieur + poids_superieur  # Poids total pour la moyenne pondérée
                moyenne_ponderee_ouverture = (poids_superieur * ouverture_inferieure + poids_inferieur * ouverture_superieure) / poids_total  # Calcul de la moyenne pondérée des ouvertures
                nouvelles_donnees.append([temp, moyenne_ponderee_ouverture])  # Ajoute la température et la moyenne pondérée d'ouverture aux nouvelles données
            elif temp_inferieure:
                nouvelles_donnees.append([temp, temp_a_ouverture[temp_inferieure][-1]])  # Ajoute la température et la dernière ouverture disponible à la température inférieure
            elif temp_superieure:
                nouvelles_donnees.append([temp, temp_a_ouverture[temp_superieure][0]])  # Ajoute la température et la première ouverture disponible à la température supérieure
    return nouvelles_donnees  # Retourne les données avec les températures manquantes remplies par interpolation linéaire

# Fonction principale pour effectuer l'interpolation linéaire des données
def lineariser_donnees(donnees):
    plage_temp = trouver_plage_temperature(donnees)  # Trouve la plage de température dans les données
    min_temp = min(plage_temp)  # Température minimale dans la plage
    max_temp = max(plage_temp)  # Température maximale dans la plage
    temp_a_ouverture = creer_dictionnaire_temp_to_opening(donnees)  # Crée un dictionnaire de température à ouverture
    nouvelles_donnees = remplir_temp_manquantes(plage_temp, temp_a_ouverture, min_temp, max_temp)  # Remplit les températures manquantes par interpolation linéaire
    return nouvelles_donnees  # Retourne les données avec les températures manquantes remplies
```

### Génération des paramètres nyctinastiques spécifiques à une période

```python
import datetime

def est_jour_dans_plage(jour, mois, annee, debut_jour, debut_mois, debut_annee, fin_jour, fin_mois, fin_annee):
    try:
        date_donnee = datetime.datetime(annee, mois, jour)
        date_debut = datetime.datetime(debut_annee, debut_mois, debut_jour)
        date_fin = datetime.datetime(fin_annee, fin_mois, fin_jour)
        
        return date_debut <= date_donnee <= date_fin
    except ValueError:
        return False
    
import datetime

def additionner_jours(jour_initial, mois_initial, jours_a_ajouter):
    date_initiale = datetime.datetime(year=datetime.datetime.now().year, month=mois_initial, day=jour_initial)
    nouvelle_date = date_initiale + datetime.timedelta(days=jours_a_ajouter)
    jour = nouvelle_date.day
    mois = nouvelle_date.month
    return (jour, mois)

def ajouter_jour_et_obtenir_mois(annee, mois, jour, ajout):
    try:
        date_actuelle = datetime.datetime(annee, mois, jour)
        date_suivante = date_actuelle + datetime.timedelta(days=ajout)
        mois_suivant = date_suivante.month
        
        return mois_suivant
    except ValueError:
        print("Date invalide ! Veuillez entrer une date valide.")
        

def calcul_taux_ouverture(ov_t, tp_t):
    return round(((ov_t + tp_t) / 2), 2)

# Retourne les paramètres nyctinastiques d'une période donée (simulation environnementale)
def nyctinastie_periode(periode, donnees_temperature, ov_nyctinastie, tp_nyctinastie, jf=False, debut_fleuraison=(7, 6), fin_fleuraison=(24, 9)):
    # ov_nyctinastie : Taux d’ouverture et vitesse de nyctinastie en fonction du temps
    # tp_nyctinastie : Taux d’ouverture en fonction de la température
    # jf : Journée froide
    # debut_fleuraison = 7 Juin, le début de la phase de développement floral par défaut
    # fin_fleuraison = 24 Septembre, la fin de la phase de développement floral par défaut
    # period = (jour, mois, heure)
    
    # retourne (calcul_taux_ouverture, vitesse, temperature_moyenne) sur 6 jours
    
    annee_courante = datetime.datetime.now().year
    jour_debut, mois_debut = debut_fleuraison
    jour_fin, mois_fin = fin_fleuraison
    
    jour_p, mois_p, heure_p = periode
    
    t_indice = 0 if not jf else 1
    
    if not est_jour_dans_plage(jour_p, mois_p, annee_courante, jour_debut, mois_debut, annee_courante, jour_fin, mois_fin, annee_courante):
        return [(0, 0, donnees_temperature[mois_p-1][t_indice]) for _ in range(6)]
    
    nyctinastie = []
    
    for i in range(6): 
        mois = ajouter_jour_et_obtenir_mois(annee_courante, mois_p, jour_p, i)
        tm_mois = donnees_temperature[mois-1][t_indice]
        
        if tm_mois <= -9 or tm_mois >= 38 :
            nyctinastie.append((-1, -1, tm_mois))
        elif heure_p < 7 or heure_p > 19:
            nyctinastie.append((0, 100, tm_mois))
        else:
            tp_taux_ouverture = [t[1] for t in tp_nyctinastie if tm_mois <= t[0]]
            tp_taux_ouverture = 0 if len(tp_taux_ouverture) == 0 else tp_taux_ouverture[0]

            H = 19 if ((heure_p < 7) or (heure_p > 19)) else heure_p
            ov_taux_ouverture = [k[1] for k in ov_nyctinastie[i] if k[0] == H]
            ov_taux_ouverture = 0 if len(ov_taux_ouverture) == 0 else ov_taux_ouverture[0]

            ov_vitesse = [k[2] for k in ov_nyctinastie[i] if k[0] == H]
            ov_vitesse = 0 if len(ov_vitesse) == 0 else ov_vitesse[0]

            nyctinastie.append( (calcul_taux_ouverture(ov_taux_ouverture, tp_taux_ouverture), round(ov_vitesse), tm_mois) )
    
    return nyctinastie
```

```python
# Retourne les paramètres nyctinastiques en fonction du temps et de la température
def nyctinastie_heure_temperature(heure, temperature, ov_nyctinastie, tp_nyctinastie):
    # retourne (calcul_taux_ouverture, vitesse) sur 6 jours
    if heure < 7 or heure > 19:
        return (0, 100, temperature)
    
    if temperature <= -9 or temperature >= 38:
        return (-1, -1, temperature)
    
    nyctinastie = []
    
    tp_taux_ouverture = [t[1] for t in tp_nyctinastie if temperature <= t[0]]
    tp_taux_ouverture = 0 if len(tp_taux_ouverture) == 0 else tp_taux_ouverture[0]
    
    for i in range(6):
        H = heure
        ov_taux_ouverture = [k[1] for k in ov_nyctinastie[i] if k[0] == H]
        ov_taux_ouverture = 0 if len(ov_taux_ouverture) == 0 else ov_taux_ouverture[0]

        ov_vitesse = [k[2] for k in ov_nyctinastie[i] if k[0] == H]
        ov_vitesse = 0 if len(ov_vitesse) == 0 else ov_vitesse[0]
        
        nyctinastie.append((calcul_taux_ouverture(ov_taux_ouverture, tp_taux_ouverture), round(ov_vitesse)))
        
    return nyctinastie
```

```python
from turtle import *
import colorsys
import numpy as np

def draw(angle, n):
    circle(n, 60)
    left(-angle)
    circle(n, 150)

def visualiser_lotus(taux_ouverture=100, vitesse_ouverture=100):
    
    ouverture = round(270 * (taux_ouverture / 100))
    vitesse = np.repeat([1,3,6,10,0],20)[int(vitesse_ouverture - 1)]
    
    try:
        tracer(5)
        pensize(3)
        speed(vitesse)
        hideturtle()
        h = 0.8657

        for i in range(ouverture):
            r, g, b = colorsys.hsv_to_rgb(h, 1, 1)
            pencolor(r, g, b)
            h += 0.0001
            goto(0, 0)
            left(90)
            forward(60)
            right(60)
            draw(90, -i)
        done()
    except Terminator:
        pass
```

```python
def nyctinastie_periode_journalier(periode, donnees_temperature, ov_nyctinastie, tp_nyctinastie, jf=False, debut_fleuraison=(7, 6), fin_fleuraison=(24, 9)):
    
    nyctinastie = {}
    
    # periode : jour, mois
    jour, mois = periode
    
    for heure in range(0,24):
        dataset = nyctinastie_periode((jour, mois, heure), donnees_temperature, ov_nyctinastie, tp_nyctinastie, jf, debut_fleuraison, fin_fleuraison)
        for i, data in enumerate(dataset, start=1):
            if not i in nyctinastie:
                nyctinastie[i] = []
            taux, vitesse, temp = data
            nyctinastie[i].append((heure, taux, vitesse, temp))
    
    return nyctinastie
```

```python
import matplotlib.pyplot as plt

def graphique_nyctinastie_journaliere(dataset, periode = (0,0), j = 0):

    heures = []
    taux_ouverture = []
    vitesse_ouverture = []
    temperature = []
    for data in dataset:
        h, t, v, t_ = data
        heures.append(h)
        taux_ouverture.append(t)
        vitesse_ouverture.append(v)
        temperature.append(t_)

    fig, ax1 = plt.subplots(figsize=(9,4))

    color = 'tab:red'
    ax1.set_xlabel('Heure')
    ax1.set_ylabel('Taux d\'ouverture (%)', color=color)
    ax1.set_ylim(0, 100)
    ax1.plot(heures, taux_ouverture, color=color, marker='o')
    ax1.tick_params(axis='y', labelcolor=color)

    ax3 = ax1.twinx()
    ax3.spines['right'].set_position(('outward', 60))
    color = 'tab:green'
    ax3.set_ylabel('Température moyenne', color=color)
    ax3.plot(heures, temperature, color=color)
    ax3.tick_params(axis='y', labelcolor=color)

    ax1.legend(['Taux d\'ouverture'], loc='upper left')
    ax3.legend(['Température moyenne du jour'], loc='upper right')
    
    jour_a, mois_a = periode
    jour_b, mois_b = additionner_jours(jour_a, mois_a, j)
    if not jour_a or not mois_a:
        plt.title('Mouvement nyctinastique observé')
    else:
        plt.title('Nyctinastie du Lotus observé le {} {}'.format(jour_b, nom_mois(mois_b)))
    plt.show()
```

```python
def nyctinastie_moyenne_mensuelle(mois, donnees_temperature, ov_nyctinastie, tp_nyctinastie, jf=False, debut_fleuraison=(7, 6), fin_fleuraison=(24, 9)):
    
    taux_total = 0
    vitesse_total = 0
    temperature_total = 0
    nb_record = 0
    
    for jour in range(1,31,6):
        for heure in range(0,24):
            dataset = nyctinastie_periode((jour, mois, heure), donnees_temperature, ov_nyctinastie, tp_nyctinastie, jf, debut_fleuraison, fin_fleuraison)
            for i, data in enumerate(dataset, start=1):
                taux, vitesse, temp = data
                
                taux_total += taux
                vitesse_total += vitesse
                temperature_total += temp
                nb_record += 1
    
    return (round(taux_total / nb_record, 2), round(vitesse_total / nb_record, 2), round(temperature_total / nb_record, 2))
```

```python
def nyctinastie_max_mensuelle(mois, donnees_temperature, ov_nyctinastie, tp_nyctinastie, jf=False, debut_fleuraison=(7, 6), fin_fleuraison=(24, 9)):
    
    taux_max = 0
    vitesse_max = 0
    temperature_max = 0
    
    for jour in range(1,31,6):
        for heure in range(0,24):
            dataset = nyctinastie_periode((jour, mois, heure), donnees_temperature, ov_nyctinastie, tp_nyctinastie, jf, debut_fleuraison, fin_fleuraison)
            for i, data in enumerate(dataset, start=1):
                taux, vitesse, temp = data
                
                taux_max = max(taux_max, taux)
                vitesse_max = max(vitesse_max, vitesse)
                temperature_max = max(temperature_max, temp)
    
    return (taux_max, vitesse_max, temperature_max)
```

```python
def nyctinastie_dans_annee(donnees_temperature, ov_nyctinastie, tp_nyctinastie, jf=False, debut_fleuraison=(7, 6), fin_fleuraison=(24, 9)):
    nyctinastie = {}
    
    for mois in range(1,13):
        data = nyctinastie_max_mensuelle(mois, donnees_temperature, ov_nyctinastie, tp_nyctinastie, jf, debut_fleuraison, fin_fleuraison)
        taux, vitesse, temp = data
        nyctinastie[mois] = ((taux, vitesse, temp))
    
    return nyctinastie
```

```python
# Activité annuelle de la nyctinastie
npannuelle = nyctinastie_dans_annee(temperatures_moyennes, ov_nyctinastie_lotus, tp_lotus_linearise)
print(npannuelle)
```

```python
import matplotlib.pyplot as plt

def afficher_evolution_annuelle_nyctinastie(moyennes_annuelles):
    # Extraire les données mensuelles
    mois = range(1, 13)
    ouverture_moyenne = []
    vitesse_moyenne = []
    temperature_moyenne = []
    
    for i in mois:
        o, v, t = moyennes_annuelles[i]
        ouverture_moyenne.append(o)
        vitesse_moyenne.append(v)
        temperature_moyenne.append(t)

    
    # Créer un seul graphique avec plusieurs axes
    fig, ax1 = plt.subplots(figsize=(12, 6))

    # Taux d'ouverture moyen
    color = 'tab:blue'
    ax1.set_xlabel('Mois de l\'année')
    ax1.set_ylabel('Taux d\'ouverture max (%)', color=color)
    ax1.set_ylim(0, 100)
    ax1.plot(mois, ouverture_moyenne, color=color, marker='o', label='Taux d\'ouverture max')
    ax1.tick_params(axis='y', labelcolor=color)


    # Température moyenne (ajouté à l'axe des y existant)
    ax3 = ax1.twinx()
    ax3.spines['right'].set_position(('outward', 60))
    color = 'tab:red'
    ax3.set_ylabel('Température moyenne (°C)', color=color)
    ax3.plot(mois, temperature_moyenne, color=color, label='Température moyenne')
    ax3.tick_params(axis='y', labelcolor=color)

    # Ajuster les étiquettes des mois sur l'axe des x
    plt.xticks(mois, ['Jan', 'Fév', 'Mar', 'Avr', 'Mai', 'Juin', 'Juil', 'Août', 'Sept', 'Oct', 'Nov', 'Déc'])

    # Ajouter les légendes pour chaque axe
    lines1, labels1 = ax1.get_legend_handles_labels()
    lines3, labels3 = ax3.get_legend_handles_labels()
    lines = lines1 + lines3
    labels = labels1 + labels3
    ax1.legend(lines, labels, loc='upper left')

    plt.title("Profil annuel de l'activité nyctinastique du Lotus - Region: Jiangsu, China ")
    plt.tight_layout()
    plt.grid(True)
    plt.show()
```
